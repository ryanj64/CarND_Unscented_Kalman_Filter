```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
UKF_data = []
UKF_lidar_data = []
UKF_radar_data = []
EKF_data = []
EKF_lidar_data = []
EKF_radar_data = []

# Extract data from UKF csv file
with open('UKF_log.csv') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    next(reader) # skip header
    UKF_data = list(reader)
    for row in UKF_data:
        # print(row)
        if row[0] == 'R':
            UKF_lidar_data.append(row)
        elif row[0] == 'L':
            UKF_radar_data.append(row)

# Extract data from EKF csv file
with open('EKF_log.csv') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    next(reader) # skip header
    EKF_data = list(reader)
    for row in EKF_data:
        # print(row)
        if row[0] == 'R':
            EKF_lidar_data.append(row)
        elif row[0] == 'L':
            EKF_radar_data.append(row)
            
# Convert to arrays
UKF_data = np.array(UKF_data) 
UKF_radar_data = np.array(UKF_radar_data) 
UKF_lidar_data = np.array(UKF_lidar_data)
EKF_data = np.array(EKF_data) 
EKF_radar_data = np.array(EKF_radar_data) 
EKF_lidar_data = np.array(EKF_lidar_data)
```

```python
font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 16,}

fig, ax = plt.subplots(ncols=1,nrows=1, figsize=(16,9))
f1 = UKF_radar_data[:,1]
t1 = t2 = np.arange(0, len(f1), dtype=np.float32)
f2 = np.zeros_like(t2)
f2.fill(7.815)
#ax.set_ylim(top=10)
ax.plot(t1, f1, 'g', t2, f2, 'r--')
ax.set_title('Radar NIS', font)
ax.set_xlabel('timestep k', font)
ax.set_ylabel('Epsilon', font)
ax.legend(['UKF Radar NIS', 'Xi Squared 0.05 (3 degrees of freedom)'])
plt.show()
```
